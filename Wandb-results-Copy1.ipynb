```python
import json
import numpy as np

def IsFinished(run):
    
    if run is None:
        return False
    
    if run.summary is None or run.state is None:
        return False
    
    if 'epoch' not in run.summary:
        return False
    
    return run.summary['epoch'] == 49 and run.state == 'finished'

datasets = ['mnist', 'svhn', 'stl', 'cifar10', 'cifar100', 'omniglot', 'reuters10k', '20news']
dataset_id = {v: i for i, v in enumerate(datasets)}

print(dataset_id)

# Append no20news

def PickFromSweep(all_runs, sweep_id):
    
    result = {}
    
    i = 0
    for run in all_runs:
        if not IsFinished(run):
            print(run.summary)
            print("State: %s" % run.state)
            continue
        if run.sweep is not None and run.sweep.name == sweep_id:
    #         print("=*=*=*=*=*=*=*=*=*=")
    #         print(run.config)
    #         print(run.summary)

    #         print(run.config['dataset'])
    #         print(dataset_id[run.config['dataset']])

    #         print(run.config['mu'])
    #         print(run.config['eta1'])
    #         print(run.config['gamma'])

            obj = {
                'meg': run.config['MuEtaGamma'],
                'tau': run.config['tau'],
                'nce': run.config['INCEvar'],
                'alp': run.config['alpha'],
                'K0b': run.config['K0BetaPair'],
                'dataset': dataset_id[run.config['dataset']]
#                 'accuracy': run.summary['accuracy']
            }
        
            print(json.dumps(obj))
            
            if json.dumps(obj) not in result:
                result[json.dumps(obj)] = []
            
            result[json.dumps(obj)].append(run.summary['accuracy'])
        
            i += 1
    
    print(i)
    
    for i in result:
        result[i] = np.asarray(result[i])
        result[i] = (result[i].mean(), result[i].std())
    
    return result
```

```python
def unpack(d):
    result = []
    for k in d:
        obj = json.loads(k)
        obj['accuracy'] = d[k][0]
        obj['stddev'] = d[k][1]
        result.append(obj)
    
    return result
        
valid_entries_list = unpack(valid_entries)
valid_entries_list
```

```python
max_acc = {}
megs = ["Mu 0.045, Eta 6.0, Gamma 1.5", "Mu 0.05 , Eta 5.0, Gamma 1.5", "Mu 0.04 , Eta 6.0, Gamma 1.5", "Mu 0.04 , Eta 5.0, Gamma 1.5"]
nces = ["INCE: UpperBound", "INCE: Exact", "INCE: Symmetrized"]

K0_beta_pairs = {
    "mnist":        [(5, 0), (7, 0)],
    "stl":          [(5, 0), (7, 0)],
    "omniglot":     [(5, 0), (7, 0)],
    "cifar100":     [(7, 0), (10,0)],
    "svhn":         [(7, 0), (10,0)],
    "cifar10":      [(10,0), (15,0)],
    "reuters10k":   [(50, 2.0 / 3.0), (50, 4.0 / 5.0)],
    "20news":       [(200, 2.0 / 3.0), (200, 4.0 / 5.0)],
}

for entry in valid_entries_list:
    # print(entry)
    s = "%s - %18s - %10s" % (megs[int(entry['meg'])], nces[int(entry['nce'])], datasets[entry['dataset']])
    if s not in max_acc:
        max_acc[s] = entry
        
    if entry['accuracy'] > max_acc[s]['accuracy']:
        max_acc[s] = entry

for k in max_acc:
    print("%s [K0-beta (%3d, %5.2f), alpha %d] - %4.2f +- %4.2f" % (
        k,
        K0_beta_pairs[datasets[max_acc[k]['dataset']]][int(max_acc[k]['K0b'])][0],
        K0_beta_pairs[datasets[max_acc[k]['dataset']]][int(max_acc[k]['K0b'])][1],
        int(max_acc[k]['alp']),
        max_acc[k]['accuracy'] * 100,
        max_acc[k]['stddev'] * 100))
```

```python
max_acc_Di = [0 for i in range(len(datasets))]
for entry in valid_entries_list:
    max_acc_Di[entry['dataset']] = max(max_acc_Di[entry['dataset']], entry['accuracy'])

max_acc_Di
```

```python
import numpy as np
tmp_list = np.zeros((4,3,3,8))
for idx in range(len(valid_entries_list)):
    entry = valid_entries_list[idx]
    m = int(entry['meg'])
    a = int(entry['alp'])
    i = int(entry['nce'])
    d = int(entry['dataset'])
    if(entry['accuracy'] > tmp_list[m,a,i,d]):
        tmp_list[m,a,i,d] = entry['accuracy']

scores = (tmp_list / np.asarray(max_acc_Di).reshape([1,1,1,8])).mean(-1)
print(scores)
```

```python
import numpy as np

results = []

for mu in ['0.1', '0.2']:
    for gamma in ['0.45', '0.55']:
        for eta1 in ['5.0', '6.0']:
            
            max_56_Di = [0 for i in range(len(datasets))]
            
            for entry in valid_entries_list:
                
                if entry['mu'] != mu or entry['gamma'] != gamma or entry['eta1'] != eta1:
                    continue
                
                max_56_Di[entry['dataset']] = max(
                    max_56_Di[entry['dataset']],
                    entry['accuracy'] / max_acc_Di[entry['dataset']]
                )
            
            avg_score = np.asarray(max_56_Di).mean()
            
            results.append('mu %s | gamma %s | eta1 %s = %f' % (mu, gamma, eta1, avg_score))
    
for r in results:
    print(r)
```
